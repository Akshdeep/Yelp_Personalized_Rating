Calculate user user similarity

```python
def getUserUserSimilarityForCategory(category, reviews_df, business_df, user_id_i, user_id_j):
    #mcp = business_df["categories"].str.contains(category, na=False).count()
    category_business = business_df[business_df["categories"].str.contains(category, na=False)]

    user_i_j_same_stars = 0
    user_i_j_same_visited = 0
    
    user_i_j_cat = reviews_df[(reviews_df["business_id"].isin(category_business['business_id']))\
                                & ((reviews_df["user_id"] == user_id_i) | (reviews_df["user_id"] == user_id_j))].groupby(["user_id","business_id"]).mean().reset_index()
    
    for row in user_i_j_cat['business_id'].unique():
        user_i_j_cat_k = user_i_j_cat[user_i_j_cat['business_id'] == row]
        user_i_j_same_stars += int(not user_i_j_cat_k.empty and user_i_j_cat_k["stars"].iloc[0] == user_i_j_cat_k["stars"].iloc[1])
        user_i_j_same_visited += int(len(user_i_j_cat_k.index) == 2)
    
    return user_i_j_same_stars * 1.0 / user_i_j_same_visited
    
    
getUserUserSimilarityForCategory("Steakhouses", review_df_train, business_df, "ymlnR8UeFvB4FZL56tCZsA", "vtRc2ApxN_qYuPDWXAMvtQ")    
    
    
```
